# Calc Corners

```python
df_FT = df[['Liga','Id_Jogo','Date','Home','Away','FT_Cantos_H','FT_Cantos_A','FT_Cantos_Total','Odds_Corners_H','Odds_Corners_D','Odds_Corners_A','Odds_Corners_Over75','Odds_Corners_Over85','Odds_Corners_Over95','Odds_Corners_Over105','Odds_Corners_Over115']]

# Média
df_FT['Media_Cantos_Marcados_Home'] = df_FT.groupby('Home')['FT_Cantos_H'].rolling(window=10, min_periods=1).mean().reset_index(0,drop=True)
df_FT['Media_Cantos_Sofridos_Home'] = df_FT.groupby('Home')['FT_Cantos_A'].rolling(window=10, min_periods=1).mean().reset_index(0,drop=True)
df_FT['Media_Cantos_Total_Home'] = df_FT.groupby('Home')['FT_Cantos_Total'].rolling(window=10, min_periods=1).mean().reset_index(0,drop=True)

# Desvio Padrão
df_FT['DesvP_Cantos_Marcados_Home'] = df_FT.groupby('Home')['FT_Cantos_H'].rolling(window=10, min_periods=1).std().reset_index(0,drop=True)
df_FT['DesvP_Cantos_Sofridos_Home'] = df_FT.groupby('Home')['FT_Cantos_A'].rolling(window=10, min_periods=1).std().reset_index(0,drop=True)
df_FT['DesvP_Cantos_Total_Home'] = df_FT.groupby('Home')['FT_Cantos_Total'].rolling(window=10, min_periods=1).std().reset_index(0,drop=True)

# Coeficiente de Variação
df_FT['CV_Cantos_Marcados_Home'] = df_FT['DesvP_Cantos_Marcados_Home'] / df_FT['Media_Cantos_Marcados_Home']
df_FT['CV_Cantos_Sofridos_Home'] = df_FT['DesvP_Cantos_Sofridos_Home'] / df_FT['Media_Cantos_Sofridos_Home']
df_FT['CV_Cantos_Total_Home'] = df_FT['DesvP_Cantos_Total_Home'] / df_FT['Media_Cantos_Total_Home']

# Média
df_FT['Media_Cantos_Marcados_Away'] = df_FT.groupby('Away')['FT_Cantos_A'].rolling(window=10, min_periods=1).mean().reset_index(0,drop=True)
df_FT['Media_Cantos_Sofridos_Away'] = df_FT.groupby('Away')['FT_Cantos_H'].rolling(window=10, min_periods=1).mean().reset_index(0,drop=True)
df_FT['Media_Cantos_Total_Away'] = df_FT.groupby('Away')['FT_Cantos_Total'].rolling(window=10, min_periods=1).mean().reset_index(0,drop=True)

# Desvio Padrão
df_FT['DesvP_Cantos_Marcados_Away'] = df_FT.groupby('Away')['FT_Cantos_A'].rolling(window=10, min_periods=1).std().reset_index(0,drop=True)
df_FT['DesvP_Cantos_Sofridos_Away'] = df_FT.groupby('Away')['FT_Cantos_H'].rolling(window=10, min_periods=1).std().reset_index(0,drop=True)
df_FT['DesvP_Cantos_Total_Away'] = df_FT.groupby('Away')['FT_Cantos_Total'].rolling(window=10, min_periods=1).std().reset_index(0,drop=True)

# Coeficiente de Variação
df_FT['CV_Cantos_Marcados_Away'] = df_FT['DesvP_Cantos_Marcados_Away'] / df_FT['Media_Cantos_Marcados_Away']
df_FT['CV_Cantos_Sofridos_Away'] = df_FT['DesvP_Cantos_Sofridos_Away'] / df_FT['Media_Cantos_Sofridos_Away']
df_FT['CV_Cantos_Total_Away'] = df_FT['DesvP_Cantos_Total_Away'] / df_FT['Media_Cantos_Total_Away']
```

```python
df_FT = df_FT[['Liga','Date','Id_Jogo','Home','Away','FT_Cantos_H','FT_Cantos_A','FT_Cantos_Total',
 'Media_Cantos_Marcados_Home','CV_Cantos_Marcados_Home','Media_Cantos_Sofridos_Home','CV_Cantos_Sofridos_Home',
 'Media_Cantos_Total_Home','CV_Cantos_Total_Home',
 'Media_Cantos_Marcados_Away','CV_Cantos_Marcados_Away','Media_Cantos_Sofridos_Away','CV_Cantos_Sofridos_Away',
 'Media_Cantos_Total_Away','CV_Cantos_Total_Away','Odds_Corners_H','Odds_Corners_D','Odds_Corners_A','Odds_Corners_Over75','Odds_Corners_Over85','Odds_Corners_Over95','Odds_Corners_Over105','Odds_Corners_Over115']]
df_FT
```

```python
df_FT.dropna(inplace = True)
# Ajustando o Índice
df_FT.reset_index(inplace=True, drop=True)
df_FT.index = df_FT.index.set_names(['Nº'])
df_FT = df_FT.rename(index=lambda x: x + 1)
df_FT
```

```python
df_FT_Home = df_FT[['Liga','Date','Home','Media_Cantos_Marcados_Home','CV_Cantos_Marcados_Home',
                    'Media_Cantos_Sofridos_Home','CV_Cantos_Sofridos_Home',
                    'Media_Cantos_Total_Home','CV_Cantos_Total_Home','Odds_Corners_H','Odds_Corners_D','Odds_Corners_A','Odds_Corners_Over75','Odds_Corners_Over85','Odds_Corners_Over95','Odds_Corners_Over105','Odds_Corners_Over115']]
df_FT_Home = df_FT_Home.groupby('Home').tail(1)
df_FT_Home = df_FT_Home.sort_values(by=['Home'])
df_FT_Home.reset_index(inplace=True, drop=True)
df_FT_Home.index = df_FT_Home.index.set_names(['Nº'])
df_FT_Home = df_FT_Home.rename(index=lambda x: x + 1)
df_FT_Home
```

```python
df_FT_Away = df_FT[['Liga','Date','Away','Media_Cantos_Marcados_Away','CV_Cantos_Marcados_Away',
                    'Media_Cantos_Sofridos_Away','CV_Cantos_Sofridos_Away',
                    'Media_Cantos_Total_Away','CV_Cantos_Total_Away','Odds_Corners_H','Odds_Corners_D','Odds_Corners_A','Odds_Corners_Over75','Odds_Corners_Over85','Odds_Corners_Over95','Odds_Corners_Over105','Odds_Corners_Over115']]
df_FT_Away = df_FT_Away.groupby('Away').tail(1)
df_FT_Away = df_FT_Away.sort_values(by=['Away'])
df_FT_Away.reset_index(inplace=True, drop=True)
df_FT_Away.index = df_FT_Away.index.set_names(['Nº'])
df_FT_Away = df_FT_Away.rename(index=lambda x: x + 1)
df_FT_Away
```

```python
import pandas as pd
from datetime import datetime

# Assuming you have DataFrames La_Liga_Games, Premier_Games, and Brasil_Games

# Get the current datetime
current_datetime = datetime.now()

print(current_datetime)
# Find the first date after today for each league
first_date_after_today_la_liga = La_Liga_Games.loc[La_Liga_Games['Date'] > current_datetime, 'Date'].min()
first_date_after_today_premier_league = Premier_Games.loc[Premier_Games['Date'] > current_datetime, 'Date'].min()
first_date_after_serie_a = Brasil_Games.loc[Brasil_Games['Date'] > current_datetime, 'Date'].min()

# Filter rows for each league based on the first date after today
filtered_la_liga = La_Liga_Games[(La_Liga_Games['Date'] == first_date_after_today_la_liga)]
filtered_premier_league = Premier_Games[(Premier_Games['Date'] == first_date_after_today_premier_league)]
filtered_serie_a = Brasil_Games[(Brasil_Games['Date'] == first_date_after_serie_a)]
```

```python
df_round = pd.concat([filtered_la_liga,filtered_premier_league,filtered_serie_a])
# setting pandas print options
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.set_option('display.max_colwidth', 100)
df_round.rename(columns={'Date': 'Date_round'}, inplace=True)
df_round.head()
```

```python
merged_df_Home = df_round.merge(df_FT_Home, on='Home', how='inner')
merged_df_Home = merged_df_Home[['Date_round', 'League', 'Round', 'Home', 'Media_Cantos_Marcados_Home',
                               'CV_Cantos_Marcados_Home', 'Media_Cantos_Sofridos_Home',
                               'CV_Cantos_Sofridos_Home']]
merged_df_Home.head()
```

```python
merged_df_Away = df_round.merge(df_FT_Away, on='Away', how='inner')
merged_df_Away = merged_df_Away[['Date_round','League','Round','Away','Media_Cantos_Marcados_Away',
                                 'CV_Cantos_Marcados_Away','Media_Cantos_Sofridos_Away',
                                 'CV_Cantos_Sofridos_Away'
]]
merged_df_Away.head()
```

```python
merged_df = df_round.merge(df_FT_Away, on='Away').merge(df_FT_Home, on='Home')
merged_df = merged_df[['Date_round','League','Round','Home','Media_Cantos_Marcados_Home',
                               'CV_Cantos_Marcados_Home','Media_Cantos_Sofridos_Home',
                               'CV_Cantos_Sofridos_Home','Away','Media_Cantos_Marcados_Away',
                                 'CV_Cantos_Marcados_Away','Media_Cantos_Sofridos_Away',
                                 'CV_Cantos_Sofridos_Away'
]]
merged_df
```

```python
df_Final = merged_df[(merged_df.CV_Cantos_Marcados_Home <= 0.5) & (merged_df.CV_Cantos_Marcados_Away <= 0.5)]
df_Final
```

```python
df_Final['Corner_Home'] = df_Final.apply(lambda row: min(row['Media_Cantos_Marcados_Home'], row['Media_Cantos_Sofridos_Away']), axis=1)

# Calcular Corner_Away
df_Final['Corner_Away'] = df_Final.apply(lambda row: min(row['Media_Cantos_Marcados_Away'], row['Media_Cantos_Sofridos_Home']), axis=1)

# Calcular Corner_Bet
df_Final['Corner_Bet'] = df_Final['Corner_Home'] + df_Final['Corner_Away']

df_Final.head()
```
